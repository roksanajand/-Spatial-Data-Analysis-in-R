# a) Create a point set with a Poisson distribution in a window with a radius of 10.
Define the intensity of the process.


```r

# Creating the window (circle) with radius 10
window <- disc(radius=10)

# Generating points based on a Poisson process with intensity 0.9
poisson_model <- rpoispp(0.9, win = window)

# Plotting the Poisson process
plot(poisson_model, main = "Point Set with Poisson Distribution in a Window with Radius 10")
```

# b) Generating a regular Strauss point process
 Generate points based on the Strauss process with intensity 0.8, 
interaction between points 0.5, and interaction coefficient 1.


```r

# Generating the Strauss process with intensity 0.8, interaction radius 0.5, and interaction coefficient 1
strauss_model <- rStrauss(0.8, 0.5, 1, window)

# Plotting the Strauss process
plot(strauss_model, main = "p_regular - Strauss Process")
```

# c) Generating a clustered Matern point process
 Generate points based on the Matern process with intensity 0.2,
cluster dispersion parameter 2, and mean number of points per cluster 10.


```r
# Generating the Matern process with intensity 0.2, cluster dispersion 2, and mean points per cluster 10
cluster_model <- rMatClust(0.2, 2, 10, window)

# Plotting the Matern process
plot(cluster_model, main = "p_cluster - Matern Process")

```

```r
# Comparison of processes - visualization on one plot
# Divide the window into 3 plots side by side
par(mfrow = c(1, 3))

# Plotting the Poisson process
plot(poisson_model, main = "ppois - Poisson Process")

# Plotting the Strauss process
plot(strauss_model, main = "p_regular - Strauss Process")

# Plotting the Matern process
plot(cluster_model, main = "p_cluster - Matern Process")

# Reset to a single plot window
par(mfrow = c(1, 1))
```

# Task 2
Calculating the distance of each point to the nearest neighbor


```r
# For the Poisson process
poisson_nndist <- nndist(poisson_model)

# For the Strauss process
strauss_nndist <- nndist(strauss_model)

# For the Matern process
cluster_nndist <- nndist(cluster_model)

# Visualization of histograms of distances
# Dividing the window into 3 side-by-side plots
par(mfrow = c(1, 3))

# Histogram for the Poisson process
hist(poisson_nndist, main = "ppois - Poisson Process", 
     xlab = "Distance", 
     ylab = "Frequency", 
     col = "lightblue")

# Histogram for the Strauss process
hist(strauss_nndist, main = "p_regular - Strauss Process", 
     xlab = "Distance", 
     ylab = "Frequency", 
     col = "lightgreen")

# Histogram for the Matern process
hist(cluster_nndist, main = "p_cluster - Matern Process", 
     xlab = "Distance", 
     ylab = "Frequency", 
     col = "lightcoral")

# Reset to a single plot window
par(mfrow = c(1, 1))
```

# Task 3
Calculating the G-function for each process
The G-function calculates the probability that a point has its nearest neighbor at distance r.
Compute the G-function values for all three processes and display the G-function values depending on r.


```r

# G-function for the Poisson process
G_poisson <- Gest(poisson_model)

# G-function for the Strauss process
G_strauss <- Gest(strauss_model)

# G-function for the Matern process
G_cluster <- Gest(cluster_model)

# Visualization of the G-function
# Plotting the G-function for the Poisson process
par(mfrow = c(1, 1))
plot(G_poisson, main = "G-function for Poisson", col = "blue")

# Adding the G-function for the Strauss process
plot(G_strauss, add = TRUE, col = "green")

# Adding the G-function for the Matern process
plot(G_cluster, add = TRUE, col = "red")

# Adding a legend to the plot
legend("topright", legend = c("Poisson", "Regular - Strauss", "Clustered - Matern"), col = c("blue", "green", "red"), lty = 1)
```

# Task 4
Calculating the K-function for each process


```r
# The K-function for the Poisson process
K_poisson <- Kest(poisson_model, correction = "border")

# The K-function for the Strauss process
K_strauss <- Kest(strauss_model, correction = "border")

# The K-function for the Matern process
K_cluster <- Kest(cluster_model, correction = "border")

# Visualization of the K-function
# Plotting the K-function for the Poisson process
plot(K_poisson, main = "K-function for different processes", 
     xlab = "Distance r", 
     ylab = "K(r)", 
     col = "blue", 
     lwd = 2)

# Adding the K-function for the Strauss process
plot(K_strauss, add = TRUE, col = "green", lwd = 2)

# Adding the K-function for the Matern process
plot(K_cluster, add = TRUE, col = "red", lwd = 2)

# Adding a legend to the plot
legend("bottomright", legend = c("p_poisson", "Regular - Strauss", "Clustered - Matern"),
       col = c("blue", "green", "red"), lwd = 2)
```

# Task 5
Calculating distance maps for each process


```r
# Distance map for the Poisson process
distmap_poisson <- distmap(poisson_model)

# Distance map for the Strauss process
distmap_strauss <- distmap(strauss_model)

# Distance map for the Matern process
distmap_cluster <- distmap(cluster_model)

# Visualization of distance maps
# Dividing the window into 3 plots side by side
par(mfrow = c(1, 3))

# Distance map for the Poisson process
plot(distmap_poisson, main = "p_poisson")

# Distance map for the Strauss process
plot(distmap_strauss, main = "Regular - Strauss")

# Distance map for the Matern process
plot(distmap_cluster, main = "Clustered - Matern")
```
